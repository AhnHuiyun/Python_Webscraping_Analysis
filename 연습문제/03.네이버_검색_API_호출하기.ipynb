### 3-1. 네이버 책 검색 API 호출하기

```python
import requests
import json
from datetime import datetime
import pandas as pd

CLIENT_ID = 'J8fnyyTMFt1eHHqf_zXk'
CLIENT_SECRET = 'VBvhidVVgn'

headers = {
    'X-Naver-Client-Id': CLIENT_ID,
    'X-Naver-Client-Secret': CLIENT_SECRET
}

def search_books(query):
    url = 'https://openapi.naver.com/v1/search/book.json'
    params = {
        'query': query,
        'display': 30,
        'sort': 'sim'
    }
    res = requests.get(url, params=params, headers=headers)
    return res.json()['items']

def save_books_to_json(books, filename='../data/books.json'):
    with open(filename, 'w', encoding='utf-8') as f:
        json.dump(books, f)

def load_books_to_dataframe(json_path='../data/books.json'):
    book_df = pd.read_json(json_path)
    return book_df

def preprocess_dataframe(df):
    df = df.copy()
    
    df['description'] = df['description']
    
    return df[['title', 'link', 'image', 'author', 'discount', 
              'publisher', 'pubdate', 'isbn', 'description']]

books = search_books("파이썬")
```

#### 1. 질문 :  검색어로  찾은  책 목록을 json 파일로 저장하기

```python
save_books_to_json(books)
```

#### 2. books.json 파일을 Pandas DataFrame로 저장하기

```python
book_df = load_books_to_dataframe()
```

#### 3. 질문 :  검색어로  찾은  책 목록 출력하기

```python
processed_df = preprocess_dataframe(book_df)
processed_df.head()
```

```python
book_df = book_df[book_df['discount'] >= 20000]

book_df_column = book_df[['title', 'author', 'discount', 'publisher', 'pubdate']]
book_df_column_sort = book_df_column.sort_values(by='discount', ascending=False).reset_index(drop=True)

book_df_column_sort.head()
```

### 3-2. 네이버 쇼핑 검색 API 호출하기

```python
import requests
import json
import pandas as pd

CLIENT_ID = 'J8fnyyTMFt1eHHqf_zXk'
CLIENT_SECRET = 'VBvhidVVgn'

headers = {
    'X-Naver-Client-Id': CLIENT_ID,
    'X-Naver-Client-Secret': CLIENT_SECRET
}

def search_shops(query):
    params = {
        'query': query,
        'display': 50,
        'sort': 'sim'
    }

    url = 'https://openapi.naver.com/v1/search/shop.json'

    res = requests.get(url, params=params, headers=headers)
    return res.json()['items']

def save_shops_to_json(data, filename='../data/shops.json'):
    with open(filename, 'w', encoding='utf-8') as f:
        json.dump(data, f)

def load_shops_to_dataframe(filename='../data/shops.json'):
    return pd.read_json(filename)

shop_data = search_shops('가디건')
```

#### 1. shops.json 파일을 Pandas DataFrame로 저장하기

```python
save_shops_to_json(shop_data)

shop_df = load_shops_to_dataframe()
```

#### 2. 질문 :  검색어로  찾은  Shop의 상품  목록 출력하기

```python
shop_df_column = [
    'title', 'link', 'image', 'lprice', 'hprice', 'mallName',
    'productId', 'productType', 'brand', 'maker',
    'category1', 'category2', 'category3', 'category4'
]

shop_df[shop_df_column].head()
```

#### 3. 질문 :  검색어로  찾은  Shop의 상품  목록 중에서 가격이 50,000원 이하인 상품만 출력하기

```python
shop_df = shop_df[shop_df['lprice'] <= 50000]
shop_df_column = shop_df[['brand', 'lprice', 'mallName', 'link']]
shop_df_column_sort = shop_df_column.sort_values(by='lprice', ascending=True).reset_index(drop=True)

shop_df_column_sort.head()
```

#### 4. 질문 :  검색어로  찾은  Shop의 상품  목록 중에서 특정 쇼핑몰 상품만 출력하기

```python
shop_df = shop_df[shop_df['mallName'] == '네이버']

shop_df_column = shop_df.loc[:, 'lprice':'brand']
shop_df_column_sort = shop_df_column.sort_values(by='lprice', ascending=True).reset_index(drop=True)

shop_df_column_sort.head()
```
